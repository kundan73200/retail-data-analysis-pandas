```python
#----------Fixing Date-----
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
#-----------counting upi-------
#df['upi']=df.loc[df['payment_method']=="UPI",'payment_method'].count()
#df
df=df[["order_id","customer_id","product_id","order_date","status","payment_method","city","price","total_amount"]]
```

```python
#-----------Average from each------------------

average_df =df.groupby('payment_method')['total_amount'].agg(['sum','mean']).reset_index()
#average_df2 =df.groupby('payment_method')['total_amount'].mean().rename('average_amount')

#new_dff=pd.merge(df,average_df,average_df2,on='payment_method')
average_df.columns= ['payment_method','total_amount','average_amount']
average_df






#-----Sheperate  sheet for only Payments details----

 #-----All Paymets----------
df1=df[['customer_id','order_id','payment_method','total_amount']]
df1
upi_df1=df1



#------UPI Details------------------

df2=df[df['payment_method']=="UPI"]
df2=df2[['order_id','total_amount']]
upi_df2=df2

#---------Credit card Details------------
df3=df[df['payment_method']=="Credit Card"]
df3=df3[["order_id","payment_method",'total_amount']]
upi_df3=df3

#----------Debit Card Details---------------
df4=df[df['payment_method']=="Debit Card"]
df4=df4[["order_id","payment_method",'total_amount']]
upi_df4=df4


#------------Total Income----------------
#Payment_df= df.groupby('payment_method')['total_amount'].sum()
#payment_df.loc[len(payment_df)]=['Total',payment_df['total_amount'].sum()]
```

```python
    #--------top selling amounts----------

df['months']=df['order_date'].dt.to_period('M')
df_monthly_sale=df.groupby('months')['total_amount'].sum().reset_index()

df_high_monthly_sale=df_monthly_sale.sort_values(by="total_amount",ascending=False).reset_index(drop=True)

df_high_monthly_top10=df_high_monthly_sale.loc[0:10,["months","total_amount"]]
df_high_monthly_top10=df_high_monthly_top10.rename(columns={"total_amount":"Heighest_selling_amount",
                                      "months":"top_selling_months"
                                     
                                     })
df_monthly_sale_average=df.groupby(by="months")["total_amount"].mean().reset_index()
df_monthly_sale_average=df_monthly_sale_average.sort_values(by="total_amount",ascending=False).reset_index(drop=True)

#merged_df=pd.merge(df_monthly_sale,df_monthly_sale_average,on='months',how='outer')
#all_payment_info=pd.merge(merged_df,df_high_monthly_top10,left_on='months',right_on='top_selling_months',how='outer')

#df=df[["order_id","customer_id","product_id","order_date","status","payment_method","city","price","total_amount"]]
#df_high_monthly_top10
#df_high_monthly_sale


#------top selling product---------

top_products=df.groupby('product_id')['total_amount'].sum().reset_index()
top_products=top_products.sort_values(by="total_amount",ascending=False).reset_index(drop=True)
top10_selling_products=top_products.loc[0:9,['product_id','total_amount']]
#top_products
#top10_selling_products
```

```python
#-------------top_10_customer_____
top10_customer=df.groupby("customer_id")['total_amount'].sum().reset_index()
top10_customer=top10_customer.sort_values(by="total_amount",ascending=False).head(10)
top10_customer=top10_customer.reset_index(drop=True)
top10_customer
df
```

```python
#-----------Inique Customers
df_unique_customer=pd.DataFrame(df['customer_id'].unique(),columns=['customer_id'])
df_unique_customer

#-----------repet_customers--------------
repeat_customer=df.groupby('customer_id')['order_id'].count().reset_index()
repeat_customer=repeat_customer[repeat_customer['order_id']>1]
repeat_customer.rename(columns={'order_id':'order_count'},inplace=True)
repeat_customer=repeat_customer.sort_values(by='order_count',ascending=False)
print(repeat_customer)
```

```python
#-----------common Order Status-------------
df_status_count=df['status'].value_counts().reset_index()
df_status_count.columns=['status','status_count']
df_status_count=df_status_count.sort_values(by="status_count",ascending=False)
df_status_count
```

```python
#-----------cancelation rate
df_total_order=df['order_id'].count()
df_cancelled_order=df[df['status']=="Cancelled"]['status'].count()
cancellation_rate=(df_cancelled_order/df_total_order)*100
cancellation_rate=round(cancellation_rate,2)

cancel_df=pd.DataFrame({'cancellation_rate':[cancellation_rate]})


cancel_df
```

```python
#-------sale by city-----------
df_top_selling_city=df.groupby('city')['total_amount'].sum().reset_index()

df_top_performing_city=df_top_selling_city.sort_values(by="total_amount",ascending=False).reset_index()
df_top_performing_city
df_top_selling_city







#-----Creating Sheets-------------


with pd.ExcelWriter("Payment.xlsx",engine='openpyxl') as writer:
    upi_df1.to_excel(writer,sheet_name="All Payment",index=False)
    upi_df2.to_excel(writer,sheet_name="Upi info",index=False)
    upi_df3.to_excel(writer,sheet_name="Credit card Details",index=False)
    upi_df4.to_excel(writer,sheet_name="Debit card Details",index=False)
    average_df.to_excel(writer,sheet_name="Average Amounts",index=False)
    df_monthly_sale.to_excel(writer,sheet_name="Monthly_selling",index=False)
    df_monthly_sale_average.to_excel(writer,sheet_name="Average sale for months",index=False)
    df_high_monthly_top10.to_excel(writer,sheet_name="top Sellings Months",index=False)
    top_products.to_excel(writer,sheet_name="top_selling_products",index=False)
    top10_selling_products.to_excel(writer,sheet_name="top10_selling_products",index=False)
    top10_customer.to_excel(writer,sheet_name="top10_customer",index=False)
    df_unique_customer.to_excel(writer,sheet_name="df_unique_customer",index=False)
    df_unique_customer.to_excel(writer,sheet_name="df_unique_customer",index=False)
    repeat_customer.to_excel(writer,sheet_name="repeat_customers",index=False)
    df_status_count.to_excel(writer,sheet_name="df_status_count",index=False)
    cancel_df.to_excel(writer,sheet_name="cancelation_rate",index=False)
    df_top_selling_city.to_excel(writer,sheet_name="top_selling_city",index=False)
    df_top_performing_city.to_excel(writer,sheet_name="df_top_performing_city",index=False)
    
```
